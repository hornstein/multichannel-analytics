# Visualisering

Den här notebooken innehåller ett antal exempel på hur man kan visualisera data för uppgift 1 i Jupyter notebook.

```
from google.colab import drive
drive.mount('/content/drive')
```

```
import pandas as pd

combined_df = pd.read_csv('/content/drive/My Drive/combined_df.csv')
combined_df.head()
```

```
import seaborn as sns

sns.lineplot(x='start_date', y='transactionrevenue', hue='campaign', data=combined_df);
```

```
sns.pairplot(combined_df[['campaign', 'sessions',	'pageviews', 'transactions', 'transactionrevenue']], hue='campaign', kind="reg");
```

```
sns.barplot(x="campaign", y="transactionrevenue", hue="Gender", data=combined_df);
```

```
age_category = pd.cut(combined_df.Age,bins=[0,18,25,35,45,55,65],labels=['-18','18-25','25-35','35-45','45-55','55-65'])

sns.barplot(x="campaign", y="transactionrevenue", hue=age_category, data=combined_df);
```

```
sns.catplot("campaign", "Age", "Gender", data=combined_df, kind="box");
```

```
import matplotlib.pyplot as plt

sns.distplot(combined_df.loc[combined_df['campaign']=='black_friday']['Age'], label='black_friday')
sns.distplot(combined_df.loc[combined_df['campaign']=='superdupersale']['Age'], label='superdupersale')
sns.distplot(combined_df.loc[combined_df['campaign']=='winter_getaways']['Age'], label='winter_getaways')
sns.distplot(combined_df.loc[combined_df['campaign']=='cyber_monday']['Age'], label='cyber_monday')

plt.legend();
```

```
sns.violinplot(x = 'campaign', y = 'pageviews', data = combined_df);
```
